```python
import requests
import pandas as pd
import numpy as np
import time
import random
from bs4 import BeautifulSoup as soup
import lxml
import re
import os
```

```python
root_url = 'https://coinmarketcap.com'

path = root_url + '/exchanges/volume/24-hour/all/'

response = requests.get(path)
da = soup(response.content,  'html.parser')
raw_data = da.find("table", {"class": "table"})
raw_data
```

```python
root_url = 'https://coinmarketcap.com'
url = '/exchanges/upbit/'
path_ex = root_url + url
overall_table = len(pd.read_html(path_ex)[0])
# response = requests.get(path)
# indi_exchange = soup(response.content,  'html.parser')
# # raw_data = da.find("table", {"class": "table"})
```

```python
ex_vol = pd.DataFrame(columns=['exchange','total_volume','NumCoin'])

ex = ''
vol = ''
url = ''
for item in raw_data.select('tr'):
    for indi_item in item.select('h3'):
        if indi_item:
            ex = indi_item.get_text().split('. ')[-1]
    for indi_item in item.select('td.text-center a'):
        if indi_item:
            url=indi_item.get('href')
    for indi_item in item.select('td.bold.volume.text-right'):
        if indi_item:
            vol = indi_item.get('data-usd')
            row = {}
            row['exchange'] = ex
            row['total_volume'] = vol
            time.sleep(1)
            row['NumCoin'] = len(pd.read_html(root_url + url)[0])
            ex_vol = ex_vol.append(row, ignore_index=True)
```
