```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import statsmodels.api as sm
```

```python
df_red = pd.read_csv('dataset/winequality-red.csv', delimiter=';')
df_white = pd.read_csv('dataset/winequality-white.csv', delimiter=';')
```

```python
df_red.head()
```

```python
df_white.head()
```

```python
df_red['type'] = 'red'
df_white['type'] = 'white'

df_full = pd.concat([df_red,df_white],axis=0)
```

```python
df_full.head()
```

```python
df_full.columns
```

```python
quality = pd.crosstab(index=df_full["quality"], columns='count')
quality
```

```python
df_full.to_stata('wine.dta')
```

```python
df_full.groupby('quality').describe()
```

```python
X = df_full[['fixed acidity', 'volatile acidity', 'citric acid', 
             'residual sugar','chlorides', 'free sulfur dioxide', 
             'total sulfur dioxide', 'density', 'pH', 'sulphates', 
             'alcohol','type']]
y = df_full[['quality']]
X = sm.add_constant(X)

model = sm.MNLogit(y,X).fit()
# model = sm.GLM(y, X, family=sm.families.NegativeBinomial()).fit()

# predictions = model.predict(X) # make the predictions by the model
model.summary()

# mdl_margeff = model.get_margeff()
# mdl_margeff.summary()
```

```python
influence = model.get_influence()
#c is the distance and p is p-value
(c, p) = influence.cooks_distance

df_full['Cook_D'] = c
df_full['Cook_D_Pval'] = p
df_full['Cook_D_over'] = np.where(df_full['Cook_D'] > 4/len(c), 1, 0)
df_full['Cook_D_over'].sum()
```
