```python
import urllib.request as ur
from bs4 import BeautifulSoup

code =['005930']
for c in code:
    url="http://comp.fnguide.com/SVO2/ASP/SVD_Invest.asp?pGB=1&gicode=A" + c + "&cID=&MenuYn=Y&ReportGB=&NewMenuID=105&stkGb=701"
    f = ur.urlopen(url).read()
    soup=BeautifulSoup(f, 'html.parser')
    corpinfo_body = soup.find('div', class_= "section ul_corpinfo")
    hname = corpinfo_body.find('h1').string
    shcode = corpinfo_body.find('h2').string
    comparison = soup.find('table', class_= 'us_table_ty1')
print(hname, shcode, comparison)
```
