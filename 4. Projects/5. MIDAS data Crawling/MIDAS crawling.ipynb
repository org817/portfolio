```python
import requests
import pandas as pd
import time
import random
from bs4 import BeautifulSoup as soup
import lxml
import re
import os
import urllib.request
```

```python
base_url = "https://www.sec.gov/marketstructure/downloads.html"
r = requests.get(base_url)
data = r.text

soup_obj = soup(data, 'lxml')
link_dict = {}

base_table = soup_obj.select('#block-secgov-content > div > div > table > tbody > tr > td.views-field.views-field-field-display-title')
for td in base_table:
    link_dict[td.a.text] = 'https://www.sec.gov' + td.a['href']
link_dict
```

```python
for name, link in link_dict.items():
    # Create Folders
    if not os.path.exists("D:/midas/" + name):
        os.makedirs("D:/midas/" + name)
        
    r = requests.get(link).text
    soup_obj = soup(r, 'lxml')

    base_table2 = soup_obj.select('table > tbody > tr')
    download_lists = []


    for tr in base_table2:
        download_lists.append('https://www.sec.gov' + tr.td.a['href'])

    for down_indi in download_lists:
        print('download :', down_indi)
        filename = down_indi.split('/')[-1]
        urllib.request.urlretrieve(down_indi, "D:/midas/" + name + '/' + filename)
        print('end :', down_indi)
```
